```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
plt.plot(x,y,"r")#"r" =color
plt.xlabel("x(cm)")
plt.ylabel("y(cm)")
plt.title("The graph of y(cm) against x(cm)")
```

```python
plt.figure(figsize=(15,9))
plt.plot(x,y,"r")#"r" =color
plt.xlabel("x(cm)")
plt.ylabel("y(cm)")
plt.title("The graph of y(cm) against x(cm)")
```

```python
# plotting subplot 
plt.subplot(2,1,1)
plt.plot(x,y,"b")
plt.title("1st plot")
plt.subplot(2,1,2)
plt.plot(y,x,"g")
plt.title("2nd plot")
#Adding a general title to each plot
#plt.subtitle("My subplot")
```

```python
#Create figure (empty instance)
fig=plt.figure()
#Add set of axes to figure
axes=fig.add_axes([0.1,0.1,0.8,0.8])# that is left
#,bottom,width, and height range(0-1)
axes.plot(x,y,"g")
axes.set_xlabel("x(cm)")
axes.set_ylabel("y(cm)")
axes.set_title("the graph of y(cm) against x(cm)")
```

```python
#creat blank cavass 
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])#main axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3])#insert axes
#Larger figure axes1
axes1.plot(x,y,"r")
axes1.set_xlabel("x(cm)")
axes1.set_ylabel("y(cm)")
axes1.set_title("the graph of y(cm) against x(cm)")
#insert figure axes 2
axes2.plot(y,x,"g")
axes2.set_xlabel("x(cm)")
axes2.set_ylabel("y(cm)")
axes2.set_title("the graph of y(cm) against x(cm)")

```

```python
#Creating subplots
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
```

```python
#WE can iterate through an array using:
for ax in axes:
    ax.plot(x,y,"r")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title("x")
#Display the figure
fig
```

```python
#Using legend
z=x*2
#Create Figure(empty canvas)
fig=plt.figure(figsize=(8,4),dpi=100)
#add set of axes to figure
axes=fig.add_axes([0.1,0.1,0.8,0.8])#left,
#bottom,width,height (range 0 to 1)
#plot on that set of axes
axes.plot(x,y,"b",label="small")
axes.set_xlabel("x(cm)")
axes.set_ylabel("y(cm)")
axes.set_title("The graph of y(cm) against x(cm)")

axes.plot(x,z,"b",label="large")
axes.set_xlabel("x(cm)")
axes.set_ylabel("z(cm)")
axes.set_title("The graph of y(cm) against z(cm)")
axes.legend(loc=3)



```

```python
#MATLAB style line color and style
fig,ax=plt.subplots()
ax.plot(x,x**2,"b.-")#blue with dot lines
ax.plot(x,x**3,"g_")#green with dash lines

```

```python
#alpha indicates opacity
fig,ax=plt.subplots()
ax.plot(x,x+1,color="blue",alpha=0.5)#half transparent
ax.plot(x,x+2,color="#8B008B")# RGB hex code
ax.plot(x,x+3,color="#FF8C00")# RGB hex code
```
