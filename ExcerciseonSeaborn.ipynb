```python
# in the white and ticks style we can remove the top and right axis spines
#using the despine() function
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,5):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
        
sns.set_style('ticks')
sinplot()
sns.despine()#sns.despine() is only effective after the sinplot() function
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
%matplotlib inline
df=sns.load_dataset("iris")
sns.distplot(df['petal_length'],kde=False)
```

```python
# KDE means kernel density estimation and it is a way 
# to estimate the probability function of a continous variable
#by setting hist=False  in distplot yield KDE
sns.distplot(df['petal_length'],hist=False)
```

```python
# dist() is used to visualize the parametric distribution
# of a data
sns.distplot(df['petal_length'])
```

```python
# dist() is used to visualize the parametric distribution
# of a data
sns.distplot(df['petal_width'])
```

```python
#Plotting Bivariate Distribution
#jointplot function provide/ creates multi-panel figure thats
#projects d bivariate relationship between two variables
# and also the univariate distribution of each variable

sns.jointplot(x='petal_length',y='petal_width',data=df)
```

```python
# Hex bin plot:used when the data is sparse in density i.e very scatter to plot
# we pass in: kind='hex' in joint func
sns.jointplot(x='petal_length',y='petal_width',data=df,kind='hex')
```

```python
#kernel density estimation is a non-parametric way to estimate
# the distribution of a variable
# pass value 'kde' to d parameter kind
sns.jointplot(x='petal_length',y='petal_width',data=df,kind='kde')
```

```python
sns.set_style('dark')
sns.pairplot(df,hue='species',diag_kind='kde',
             kind="scatter",palette='husl')
```

```python
sns.stripplot(x='species',y='petal_length',data=df)
```

```python
#jitter adds some random noise to the data
#The parameter will arrange the positions along the categorical axis
sns.stripplot(x='species',y='petal_length',data=df,jitter=True)
```

```python
# Swarmplot() function is used as an alternate to jitter
#positions each point of scatter plot on the categorical axis 
# and thereby avoid overlapping points
sns.swarmplot(x='species',y='petal_length',data=df)
```
